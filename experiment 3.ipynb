```r
modec <- function(v){
    uniq<-unique(v)
    uniq[which.max(tabulate(match(v,uniq)))]
}
```

```r
age<-c(11,22,33,44)
weight<-c(33,456,66,755)
names<-c('hey','may','rey','jay')
print(tabulate(age))
```

```r
dataframe= data.frame(
    names,
    age,
    weight
)
df=data.frame(
    age,weight
)
print("dataframe is")
print(dataframe)
print(dataframe[1])
```

```r
summ=do.call(cbind,lapply(df,summary))
summ=rbind(summ,Mode=c(modec(dataframe$age),modec(dataframe$weight)))
print("summary")
print(summ)
```

```r
print('Correlation')
print(cor(dataframe$age,dataframe$weight))
```

```r
plot(dataframe$age,dataframe$weight,main='AGE VS WEIGHT')
```
